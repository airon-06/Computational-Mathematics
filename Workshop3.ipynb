```python
import matplotlib.pyplot as plt

def f(x):
    return 3*x + 2

x = []
y = []

for i in range(10):
    x.append(i)
    y.append(f(i))

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

i = -4*math.pi
while i <= 4*math.pi:
    x.append(i)
    y.append(math.sin(i))
    i += 0.1

plt.plot(x, y)
plt.title("y = sin(x)")
plt.grid()
plt.show()
```
