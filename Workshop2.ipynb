1.	Write a Python function that takes a list and returns a new list with distinct elements from the first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]


```python
def unique_list(lst):
    return list(set(lst))
sample=[1,2,3,3,3,3,4,5]
print(unique_list(sample))
```


2.	Write a Python function that takes a number as a parameter and checks whether the number is prime or not.


```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i==0:
            return False
    return True
print(is_prime(7))
```

3.	Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes.

```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i==0:
            return False
    return True
for i in range(3, 1000, 2):
    if is_prime(i) and is_prime(i+2):
        print((i, i+2))
```

4.	A positive integer is called a perfect number if it is equal to the sum of all of its divisors, including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors (excluding n).


```python
def sumOfDivisors(n):
    total=0
    for i in range(1, n):
        if n%i==0:
            total +=i
    return total
print(sumOfDivisors(20))
```

6.	Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.


```python
marks = [40, 70, 90, 50, 67, 52, 90]

mean = sum(marks) / len(marks)

marks.sort()
n = len(marks)

if n % 2 == 0:
    median = (marks[n//2 - 1] + marks[n//2]) / 2
else:
    median = marks[n//2]

print("Mean:", mean)
print("Median:", median)
```

8.	Write a function prodDigits() that inputs a number and returns the product of digits of that number.

```python
def prodDigits(n):
    product = 1
    for digit in str(n):
        product *= int(digit)
    return product
print(prodDigits(67))
```

9.	Write a function that will take in an unknown number of arguments and multiply all of them together and run the function for these 2 sets of numbers:
a.	1,2,3,4,5
b.	12,13,14


```python
def multiply(*args):
    result = 1
    for i in args:
        result *= i
    return result
print(multiply(1,2,3,4,5))
print(multiply(12,13,14))
```
